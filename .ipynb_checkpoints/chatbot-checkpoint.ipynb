```python
import nltk
from nltk.chat.util import Chat, reflections
```

```python
pairs = [

    # --- GREETING FLOW ---
    [
        r"hi|hello|hey|hii|heyy",
        ["Hey there! How’s your day going?"]
    ],
    [
        r"(good|fine|okay|not bad)",
        ["That’s nice to hear! What’s been the highlight of your day?"]
    ],
    [
        r"(bad|not good|sad|upset)",
        ["Oh no 😔 Want to talk about it, or should I distract you with a joke?"]
    ],

    # --- NAME FLOW ---
    [
        r"what is your name ?",
        ["I’m ChatBuddy 🤖 What’s your name?"]
    ],
    [
        r"my name is (.*)",
        ["Nice to meet you, %1! Do your friends call you something shorter?"]
    ],
    [
        r"(yes|yeah|yep)",
        ["Cool! What’s your nickname?"]
    ],
    [
        r"(no|nah)",
        ["That’s fine! %1 is already nice. What do you like doing in your free time?"]
    ],

    # --- HOBBY FLOW ---
    [
        r"my hobby is (.*)",
        ["%1 sounds fun! How did you start with that?"]
    ],
    [
        r"i like (.*)",
        ["Nice! What do you enjoy most about %1?"]
    ],

    # --- SCHOOL / WORK FLOW ---
    [
        r"i am a student",
        ["Cool! What are you studying?"]
    ],
    [
        r"i study (.*)",
        ["That’s interesting! Do you enjoy studying %1?"]
    ],
    [
        r"i am working",
        ["Awesome! What kind of work do you do?"]
    ],
    [
        r"i work as (.*)",
        ["%1 sounds like an important job! Do you like it?"]
    ],

    # --- WEEKEND FLOW ---
    [
        r"what are you doing this weekend ?",
        ["I don’t make plans, but I’d love to hear yours! Do you usually go out or stay in on weekends?"]
    ],
    [
        r"i (go out|stay home|travel|rest)",
        ["That sounds great! Do you do that often?"]
    ],

    # --- MUSIC FLOW ---
    [
        r"i like music",
        ["Music makes everything better 🎶 What’s your favorite genre?"]
    ],
    [
        r"my favorite genre is (.*)",
        ["Oh nice, %1 is a good choice! Who’s your favorite artist in that genre?"]
    ],
    [
        r"i like (.*) music",
        ["%1 music is awesome! Do you listen to it while studying or relaxing?"]
    ],

    # --- MOVIE / TV FLOW ---
    [
        r"i like movies",
        ["Same here! 🍿 What’s your favorite movie?"]
    ],
    [
        r"my favorite movie is (.*)",
        ["%1 is a great pick! Do you usually watch movies at home or in the theater?"]
    ],
    [
        r"i watch (.*)",
        ["Nice! I’ve heard good things about %1. Do you recommend it?"]
    ],

    # --- TRAVEL FLOW ---
    [
        r"i like traveling",
        ["Traveling is amazing 🌍 Where’s the best place you’ve been to?"]
    ],
    [
        r"i went to (.*)",
        ["Wow, %1 must have been fun! Would you like to visit again?"]
    ],
    [
        r"i want to visit (.*)",
        ["%1 sounds like a great destination! What attracts you there?"]
    ],

    # --- FOOD FLOW ---
    [
        r"what is your favorite food ?",
        ["I don’t eat, but pizza seems popular 🍕 What’s yours?"]
    ],
    [
        r"my favorite food is (.*)",
        ["Mmm, %1 is delicious! Do you cook it or eat out?"]
    ],
    [
        r"(pizza|burger|pasta|rice|chicken|fish)",
        ["Yum, %1! Do you have it often?"]
    ],

    # --- SPORTS FLOW ---
    [
        r"i like sports",
        ["Nice! Which sport is your favorite to play or watch?"]
    ],
    [
        r"my favorite sport is (.*)",
        ["%1 is exciting! Do you follow a specific team or player?"]
    ],

    # --- WEATHER FLOW ---
    [
        r"what’s the weather like ?",
        ["I can’t see outside, but how’s the weather where you are?"]
    ],
    [
        r"it’s (sunny|rainy|cloudy|hot|cold)",
        ["Oh, I see! Do you enjoy this kind of weather?"]
    ],

    # --- JOKE FLOW ---
    [
        r"tell me a joke",
        ["Okay! Why don’t scientists trust atoms? …Because they make up everything 😂 Want another one?"]
    ],
    [
        r"(yes|yeah|sure|ok)",
        ["Here’s one: Why was the math book sad? …Because it had too many problems 🤓"]
    ],
    [
        r"(no|nah|stop)",
        ["Alright, no more jokes 😅 What else should we talk about?"]
    ],

    # --- FUN FACT FLOW ---
    [
        r"tell me something interesting",
        ["Did you know honey never spoils? 🍯 Even 3000-year-old honey is edible! Want another fact?"]
    ],
    [
        r"(yes|yeah|sure)",
        ["Octopuses have three hearts 🐙 Crazy, right? Want more or change topic?"]
    ],
    [
        r"(no|nah|stop)",
        ["Cool! What do you want to chat about then?"]
    ],

    # --- RIDDLE FLOW ---
    [
        r"riddle",
        ["Here’s one: The more you take, the more you leave behind. What am I?"]
    ],
    [
        r"(footsteps|steps)",
        ["You got it! 🎉 Want another one?"]
    ],
    [
        r"(don’t know|give up|idk)",
        ["It was ‘footsteps’ 👣 Want another riddle or something else?"]
    ],

    # --- CREATOR FLOW ---
    [
        r"who created you ?",
        ["I was built with Python and NLTK 😃 with a little touch of Symum"]
    ],
    [
        r"(yes|yeah|a little)",
        ["That’s awesome! What kind of coding do you enjoy?"]
    ],
    [
        r"(no|nah)",
        ["That’s okay 😅 What do you like instead?"]
    ],

    # --- THANK YOU FLOW ---
    [
        r"thank you|thanks",
        ["You’re welcome 😊 What else is on your mind?"]
    ],

    # --- GOODBYE FLOW ---
    [
        r"(bye|goodbye|see you|quit)",
        ["Bye! It was nice chatting with you. Come back soon!"]
    ],

    # --- GENERIC CATCH-ALL ---
    [
        r"(.*)",
        ["That’s interesting! Tell me more.", "Oh really? Why do you think so?", "Hmm, that’s cool. What else?"]
    ],
]
```

```python
def chat():
    print("Hi, I'm a chatbot. Type 'quit' to exit.")
    chat = Chat(pairs, reflections)
    chat.converse()
    
if __name__ == "__main__":
    chat()
```
